# Data Integration

## General Services Administration

### This dataset represent time taken to hire a GSA employee from the internal request to hire through the entry on duty of the selected individual

#### Source: www.data.gov

```python
import pandas as pd
```

```python
THH1 = pd.read_csv("D:\DATA MINING\D1.csv",sep=";")
THH2 = pd.read_csv("D:\DATA MINING\D2.csv",sep=";")
```

```python
Time_to_Hire = pd.merge(THH1, THH2, on="NO")
Time_to_Hire.head(5).append(Time_to_Hire.tail(5))
```

```python
THH3 = pd.read_csv("D:\DATA MINING\D3.csv",sep=";")
```

```python
Time_to_Hire_Complete = pd.concat([Time_to_Hire, THH3])
Time_to_Hire_Complete.head(5).append(Time_to_Hire_Complete.tail(5))
```
