# Association Rule
## Market Basket Analysis
### Data : Online Retail

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv("D:\DATA MINING\online_retail.csv", sep=",")
data.shape
```

```python
data.head()
```

```python
data['Description'] = data['Description'].str.strip()
data.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
data['InvoiceNo'] = data['InvoiceNo'].astype('str')
data = data[~data['InvoiceNo'].str.contains('C')]
```

### Spain

```python
basketSpain = (data[data['Country'] =="Spain"].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))

def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket = basketSpain.applymap(encode_units)
basket.drop('POSTAGE', inplace=True, axis=1)
basket.head()
```

```python
frequent_itemsets = apriori(basket, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules[(rules['lift']>=7) & (rules['confidence']>=0.8)]
```

### Germany

```python
basketGermany = (data[data['Country'] =="Germany"].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))

basket2 = basketGermany.applymap(encode_units)
basket2.drop('POSTAGE', inplace=True, axis=1)
frequent_itemsets2 = apriori(basket2, min_support=0.05, use_colnames=True)
rules2 = association_rules(frequent_itemsets2, metric="lift", min_threshold=1)
rules2[(rules2['lift']>=3) & (rules2['confidence']>=0.5)]
```
