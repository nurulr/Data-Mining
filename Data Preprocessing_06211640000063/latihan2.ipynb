### Data Cleaning (Missing value & Outlier)

```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv("D:\DATA MINING\movie_metadata.csv",sep=";")
movie.head().append(movie.tail())
```

```python
movie.describe() #statistika deskriptif
```

```python
np.sum(movie.isnull()) #missing value
```

### missing value <10% setiap variabel diabaikan/imputasi
### missing value >15% setiap variabel dan tidak berkorelasi dg yang lain dihapus jika berkorelasi diimputasi
### missing value >50% setiap variabel dihapus variabelnya
### imputasi = mengisi missing value
#### mean (kontinu), modus (kategorik)

```python
#imputasi with mean
movie['duration'] = movie['duration'].fillna(107.201074)
movie['facenumber_in_poster'] = movie['facenumber_in_poster'].fillna(1.371446)
movie['budget'] = movie['budget'].fillna(3.975262e+07)
movie['gross'] = movie['gross'].fillna(4.846841e+07)
movie['num_user_for_reviews'] = movie['num_user_for_reviews'].fillna(272.770808)
movie['num_critic_for_reviews'] = movie['num_critic_for_reviews'].fillna(140.194272)
movie['director_facebook_likes'] = movie['director_facebook_likes'].fillna(686.621709)
movie['actor_1_facebook_likes'] = movie['actor_1_facebook_likes'].fillna(movie['actor_1_facebook_likes'].mean())
movie['actor_2_facebook_likes'] = movie['actor_2_facebook_likes'].fillna(movie['actor_2_facebook_likes'].mean())
movie['actor_3_facebook_likes'] = movie['actor_3_facebook_likes'].fillna(movie['actor_3_facebook_likes'].mean())
np.sum(movie.isnull())
```

```python
[movie['country'].mode()]
```

```python
#imputasi with mode 
from scipy.stats import mode

movie['country']=movie['country'].fillna(movie['country']).dropna().mode()[0]
np.sum(movie.isnull())
```

### Outlier

```python
#mengatasi outlier 
from scipy import stats

movie1=movie.drop('country',1)
z = np.abs(stats.zscore(movie1._get_numeric_data()))
print(z)
movie=movie[(z<3).all(axis=1)]
print(movie.shape)
```
