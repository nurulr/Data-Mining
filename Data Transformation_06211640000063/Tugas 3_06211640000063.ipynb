# Pre-processing Data Iklim di Indonesia Tahun 2011-2015
### Sumber Data : Badan Pusat Statistik (BPS)

```python
import numpy as np
import pandas as pd
import csv
```

```python
hujan1=pd.read_csv("D:\DATA MINING\Hujan.csv", sep=";")
hujan1.head()
```

```python
hujan1.describe()
```

## a. Missing value

```python
np.sum(hujan1.isnull())
```

```python
hujan1['Jumlah Curah Hujan (mm)'] = hujan1['Jumlah Curah Hujan (mm)'].fillna(hujan1['Jumlah Curah Hujan (mm)'].mean())
hujan1['Jumlah Hari Hujan (hari)'] = hujan1['Jumlah Hari Hujan (hari)'].fillna(hujan1['Jumlah Hari Hujan (hari)'].mean())
hujan1['Tekanan Udara (mb)'] = hujan1['Tekanan Udara (mb)'].fillna(hujan1['Tekanan Udara (mb)'].mean())
hujan1['Penyinaran Matahari (%)'] = hujan1['Penyinaran Matahari (%)'].fillna(hujan1['Penyinaran Matahari (%)'].mean())
hujan1['Suhu (?C)'] = hujan1['Suhu (?C)'].fillna(hujan1['Suhu (?C)'].mean())
hujan1['Kecepatan Angin (m/det)'] = hujan1['Kecepatan Angin (m/det)'].fillna(hujan1['Kecepatan Angin (m/det)'].mean())
hujan1['Kelembaban'] = hujan1['Kelembaban'].fillna(hujan1['Kelembaban'].mean())
np.sum(hujan1.isnull())
```

```python
hujan1.to_csv("hujan1.csv")
hujan1.head()
```

## b. Missing value dan Outlier

```python
from scipy import stats
hujan_=hujan1.drop(['Provinsi','Tahun'],1)
z = np.abs(stats.zscore(hujan_._get_numeric_data()))
print(z)
hujan2=hujan1[(z<3).all(axis=1)]
print(hujan2.shape)
```

```python
hujan2.to_csv("hujan2.csv")
hujan2.head()
```

## c. Missing value dan Transformasi

```python
x1=hujan1.iloc[0:,3:]
x1.head()
```

### Transformasi metode normalisasi

```python
from sklearn.preprocessing import normalize
x1norm=normalize(x1,norm='l2')
x1norm=pd.DataFrame(x1norm, columns = ['Jumlah Hari Hujan (hari)','Tekanan Udara (mb)','Penyinaran Matahari (%)','Suhu (?C)','Kecepatan Angin (m/det)','Kelembaban'])
x1norm.to_csv("x1norm.csv")
x1norm.head()
```

### Transformasi metode minmax

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,5))
scaler.fit(x1)
x1scaled=scaler.transform(x1)
x1scaled=pd.DataFrame(x1scaled, columns = ['Jumlah Hari Hujan (hari)','Tekanan Udara (mb)','Penyinaran Matahari (%)','Suhu (?C)','Kecepatan Angin (m/det)','Kelembaban'])
x1scaled.to_csv("x1scaled.csv")
x1scaled.head()
```

### Transformasi metode 3sigmoid

```python
x1sigmoidal=(1-2.718281828**(-x1norm))/(1+2.718281828**(-x1norm))
x1sigmoidal=pd.DataFrame(x1sigmoidal, columns = ['Jumlah Hari Hujan (hari)','Tekanan Udara (mb)','Penyinaran Matahari (%)','Suhu (?C)','Kecepatan Angin (m/det)','Kelembaban'])
x1sigmoidal.to_csv("x1sigmoidal.csv")
x1sigmoidal.head()
```
